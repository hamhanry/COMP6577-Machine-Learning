```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
```

```python
def generate_sample_data():
    """Generate synthetic data with clear clusters."""
    print("\n1. Generating sample data with 4 clusters...")
    X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
    return X, y

def plot_data(X, title="Sample Data"):
    """Plot the data points."""
    plt.figure(figsize=(10, 6))
    plt.scatter(X[:, 0], X[:, 1], c='gray', alpha=0.5)
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

def perform_kmeans(X, n_clusters=4):
    """Perform K-Means clustering on the data."""
    print(f"\n2. Performing K-Means clustering with {n_clusters} clusters...")
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    return kmeans

def plot_clustering_results(X, kmeans):
    """Plot the clustering results with centroids."""
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    plt.figure(figsize=(10, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', 
                s=200, linewidths=3, label='Centroids')
    plt.title('K-Means Clustering Results')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()

def find_optimal_clusters(X, max_k=10):
    """Find the optimal number of clusters using elbow method and silhouette score."""
    print("\n3. Finding optimal number of clusters...")
    inertias = []
    silhouette_scores = []
    K = range(2, max_k + 1)

    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        inertias.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(X, kmeans.labels_))

    # Plot elbow curve and silhouette scores
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(K, inertias, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')

    plt.subplot(1, 2, 2)
    plt.plot(K, silhouette_scores, 'rx-')
    plt.xlabel('k')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Analysis')

    plt.tight_layout()
    plt.show()

def iris_dataset_example():
    """Demonstrate K-Means clustering on the Iris dataset."""
    print("\n4. Working with the Iris dataset...")
    # Load Iris dataset
    iris = load_iris()
    X = iris.data

    # Perform clustering
    kmeans = KMeans(n_clusters=3, random_state=42)
    labels = kmeans.fit_predict(X)

    # Plot results (using first two features for visualization)
    plt.figure(figsize=(10, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    plt.title('K-Means Clustering on Iris Dataset (First Two Features)')
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.show()

    # Compare with actual species
    print("\nComparison with actual Iris species:")
    print("Cluster assignments:", labels[:10])
    print("Actual species:", iris.target[:10])
```

# Centroid-Based Clustering: K-Means Algorithm Tutorial

## Part 1: Synthetic Data Example

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate and plot sample data
X, y = generate_sample_data()
plot_data(X)
```

## Perform K-Means clustering

```python
kmeans = perform_kmeans(X)
plot_clustering_results(X, kmeans)

# Find optimal number of clusters
find_optimal_clusters(X)
```

## Part 2: Working with Real Data (Iris Dataset)

```python
iris_dataset_example()
```
