```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + np.random.randn(50)

# Define the learning rates to test
learning_rates = [0.001, 0.01, 0.1]

# Initialize the weights and biases
w = np.zeros((len(learning_rates),))
b = np.zeros((len(learning_rates),))

# Train the model for each learning rate
for i, lr in enumerate(learning_rates):
    w_temp = 0
    b_temp = 0
    losses = []
    for j in range(50):
        y_pred = w_temp * x + b_temp
        loss = np.mean((y_pred - y) ** 2)
        losses.append(loss)
        dw = -2 * np.mean(x * (y - y_pred))
        db = -2 * np.mean(y - y_pred)
        w_temp -= lr * dw
        b_temp -= lr * db
    w[i] = w_temp
    b[i] = b_temp

# Plot the learning rate vs. model performance
plt.plot(learning_rates, [np.mean((w[i] * x + b[i] - y) ** 2) for i in range(len(learning_rates))])
plt.xlabel('Learning Rate')
plt.ylabel('Model Performance (MSE)')
plt.title('Learning Rate vs. Model Performance')
plt.show()

# Plot the convergence of the model for each learning rate
for i, lr in enumerate(learning_rates):
    w_temp = 0
    b_temp = 0
    losses = []
    for j in range(50):
        y_pred = w_temp * x + b_temp
        loss = np.mean((y_pred - y) ** 2)
        losses.append(loss)
        dw = -2 * np.mean(x * (y - y_pred))
        db = -2 * np.mean(y - y_pred)
        w_temp -= lr * dw
        b_temp -= lr * db
    plt.plot(range(50), losses, label=f'Learning Rate: {lr}')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Convergence of the Model')
plt.legend()
plt.show()
```
